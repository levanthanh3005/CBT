```python
print("Fix CPU Col Name:")
count=0
lsCPUMetrics = ["idle","iowait","irq","nice","softirq","steal","system","user"]
lsCPUColname = []
for cn in colNames:
    if cn.find("node_cpu_seconds_total")>-1:
        name="node_cpu_seconds_total_"+lsCPUMetrics[int(count%8)]+"_cpu_"+str(int((count/8)%2))+"_"+str(int(count/16))
        print(cn," ",name)
        rareData = rareData.rename(columns={cn: name})
        lsCPUColname.append(name)
        count = count + 1
```
